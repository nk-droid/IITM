```python
import plotly.graph_objects as go
import numpy as np
```

```python
def error(w0, w1, w2, x1, x2, y):
  y_pred = -w0 + w1*x1 + w2*x2 >= 0
  error = abs(y - y_pred)
  return sum(error)
```

Q1. Generate the error surface corresponding to the following Boolean function. Take $w_0 = 1$.

What is the maximum error? (for ex, in the sample error-surface shown above, the maximum error is 4)

```python
w1 = np.linspace(-6, 6, 100)
w2 = np.linspace(-6, 6, 100)
w0 = 1
W1, W2 = np.meshgrid(w1, w2)

x1 = np.array([0,0,1,1])
x2 = np.array([0,1,0,1])
y = np.array([0,0,0,1])

errors = np.zeros((100,100))

for i in range(100):
  for j in range(100):
    errors[i,j] = error(w0, W1[i,j], W2[i,j], x1, x2, y)

fig = go.Figure(data=[go.Surface(x=W1, y=W2, z=errors)])

fig.update_layout(
    title='Error Surface',
    scene=dict(
        xaxis_title='w1',
        yaxis_title='w2',
        zaxis_title='Loss'
    )
)

fig.show()
```
